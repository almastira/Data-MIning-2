```python
import os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def load_npy(filename):
    return np.load(filename)
```

```python
dir_path = 'time_series/'
len_threshold = 1280
```

```python
X, y, ids = [], [], []

for file in os.listdir(dir_path):
    if os.path.splitext(file)[1] != '.npy':
        continue

    split = file.split("_")
    ids.append(split[0]) # track_id
    y.append(split[1][:-4]) # genre
    ts = load_npy(dir_path + file)
    
    if len(ts) > len_threshold:
        ts = ts[0:len_threshold]
    else:
        # pad = [np.mean(ts[:-5])] * (len_threshold-len(ts)) #Â fill by mean value of last n observations
        pad = [ts[-1]] * (len_threshold-len(ts)) # fill with last observation
        ts = np.append(ts, pad) 

    X.append([ts])

X, y, ids = np.array(X), np.array(y), np.array(ids)
```

```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Normalizzazione dei dati
from sklearn.preprocessing import StandardScaler

X_normalized = []

for series in X:
    scaler = StandardScaler()
    normalized_series = scaler.fit_transform(series.reshape(-1, 1)).flatten()
    X_normalized.append(normalized_series)

X_normalized = np.array(X_normalized)
```

```python
from pyts.approximation import PiecewiseAggregateApproximation
```

```python
# Assumiamo che X_normalized sia il tuo array di serie temporali normalizzate
# Definizione del numero di segmenti desiderato
n_segments = 128

# Calcolo della dimensione della finestra
window_size = len(X_normalized[0]) // n_segments

# Inizializzazione del modello PAA con window_size
paa = PiecewiseAggregateApproximation(window_size=window_size)

# Trasformazione delle serie temporali
X_paa = paa.fit_transform(X_normalized)

# X_paa Ã¨ ora l'array contenente le serie temporali trasformate
print("Dimensione originale:", X_normalized.shape)
print("Dimensione dopo PAA:", X_paa.shape)

# Esempio di visualizzazione dei risultati
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for i, (original, transformed) in enumerate(zip(X_normalized, X_paa)):
    plt.subplot(2, 1, 1)
    plt.plot(original, label=f'Original Series {i+1}' if i == 0 else "")
    plt.title("Original Time Series")
    plt.legend()

    plt.subplot(2, 1, 2)
    plt.plot(transformed, label=f'Transformed Series {i+1}' if i == 0 else "")
    plt.title("Transformed Time Series (PAA)")
    plt.legend()

    if i >= 2:  # Mostra solo le prime 3 serie per semplicitÃ 
        break

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Suddivisione del dataset in set di addestramento e di test
X_train, X_test, y_train, y_test = train_test_split(X_paa, y, test_size=0.2, shuffle=False)

# Assicurati che X_train, X_test contengano le serie temporali trasformate tramite PAA
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Calcolo del dendrogramma per ciascun metodo di linkage
linkage_methods = ['single', 'complete', 'average', 'ward']

for method in linkage_methods:
    Z = linkage(X_train_rid, method=method)
    plt.figure(figsize=(10, 5))
    dendrogram(Z)
    plt.title(f'Dendrogram (Linkage: {method})')
    plt.xlabel('Sample index')
    plt.ylabel('Distance')
    plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

clustering_methods = {
    'single': clustering1,
    'complete': clustering2,
    'average': clustering3,
    'ward': clustering3
}

for name, model in clustering_methods.items():
    labels = model.labels_
    silhouette_avg = silhouette_score(X_train_rid, labels)
    db_index = davies_bouldin_score(X_train_rid, labels)
    ch_index = calinski_harabasz_score(X_train_rid, labels)
    print(f"Method: {name}")
    print(f"Silhouette Score: {silhouette_avg}")
    print(f"Davies-Bouldin Index: {db_index}")
    print(f"Calinski-Harabasz Index: {ch_index}")
    print("-" * 30)
```

```python
Il metodo "single" ha il punteggio di silhouette piÃ¹ basso, indicando che i cluster non sono ben definiti. 
Il Davies-Bouldin Index Ã¨ relativamente basso, suggerendo una discreta compattezza dei cluster, ma il Calinski-Harabasz
Index Ã¨ il piÃ¹ basso tra i metodi testati, indicando una separazione meno buona tra i cluster.
```

```python
 Il metodo "complete" ha il punteggio di silhouette piÃ¹ alto tra i metodi testati, suggerendo cluster leggermente meglio definiti rispetto agli altri.
Il Davies-Bouldin Index Ã¨ il piÃ¹ basso, indicando la migliore compattezza dei cluster, ma il Calinski-Harabasz Index Ã¨ basso, indicando una moderata 
separazione tra i cluster.
    
```

```python
Il metodo "average" ha un punteggio di silhouette piÃ¹ basso, indicando meno definizione nei cluster. 
Tuttavia, ha il Calinski-Harabasz Index piÃ¹ alto, suggerendo una buona separazione tra i cluster, 
ma il Davies-Bouldin Index Ã¨ il piÃ¹ alto, indicando cluster meno compatti.
```

 I risultati del metodo "ward" sono identici a quelli del metodo "average", suggerendo che entrambi producono strutture di cluster molto simili.

Migliore compattezza dei cluster (Davies-Bouldin Index piÃ¹ basso): Metodo "complete".
Migliore definizione dei cluster (Silhouette Score piÃ¹ alto): Metodo "complete".
Migliore separazione tra i cluster (Calinski-Harabasz Index piÃ¹ alto): Metodo "average" e "ward" (pari).

Sulla base delle metriche fornite:

Il metodo "complete" sembra essere il migliore in termini di compattezza e definizione dei cluster.
Il metodo "average" e "ward" offrono la migliore separazione tra i cluster ma a costo di avere cluster meno compatti.
